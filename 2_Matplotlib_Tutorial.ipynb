# Matplotlib Tutorial (Beginnner Level)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Simple Matplotlib Chart

```python
x = [1,2,3]
y = [10,11,17]

plt.plot(x,y)
# Default is line chart with blue solid line
```

### Customizing Matplotlib Chart

* Color: Red - r, blue - b, Green - g
* Shape: Dashes --, Dot - o, Square - s, Triangle - ^

```python
x = [1,2,3]
y = [10,11,17]

plt.plot(x,y, 'r--')

## red dashes (r--), blue squares (bs) and green triangles(g^), red dot (ro)
## default is blue solid line (b-)
```

* Axis: axix([xmin, xmax, ymin, ymax])
* Only X-axis: xlim([xmin, xmax])
* Only Y-axis: ylim([ymin, ymax])

```python
x = [1,2,3]
y = [10,11,17]

plt.axis([0,5,9,20]) 
# axis() command takes a list of [xmin, xmax, ymin, ymax]
# Can also use: plt.xlim([0,5]) and plt.ylim([9,20])
plt.plot(x,y,'ro')
plt.show()
```

* Title: title(' ')
* X Axis Label: xlabel(' ')
* Y Axis Label: ylabel(' ')

```python
x = [1,2,3]
y = [10,11,17]


plt.title('New Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(x,y)
plt.show()
```

###  Ploting Multiple Charts in the Same Plot

```python
x1 = [1,2,3]
y1 = [10,11,17]

x2 = [5,6,7]
y2 = [14,17,19]

plt.plot(x1,y1, 'r-', linewidth=5)
plt.plot(x2,y2, 'b-', linewidth=3)

plt.title('New Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')


plt.show()
```

* Showing Grid: grid(True)

```python
x1 = [1,2,3]
y1 = [10,11,17]

x2 = [5,6,7]
y2 = [14,17,19]

plt.plot(x1,y1, 'r-', linewidth=2)
plt.plot(x2,y2, 'b-', linewidth=2)

plt.title('New Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)

plt.show()
```

* Labeling: Use label parameter in plot() function and use legend() function

```python
x1 = [1,2,3]
y1 = [10,11,17]

x2 = [5,6,7]
y2 = [14,17,19]

plt.plot(x1,y1, 'r-', linewidth=2, label="line One")
plt.plot(x2,y2, 'b-', linewidth=2, label='line Two')

plt.title('New Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0,10,10,25])
plt.legend()  # uses label parameter from plot()

plt.show()
```

### Scatter Plots

```python
x1 = [1,2,3,4,5]
y1 = [10,11,17,12,16]


x2 = [5,6,7]
y2 = [14,17,19]

plt.scatter(x1,y1,color='g',label='green')
plt.scatter(x2,y2,color='b',label='blue')

plt.title('New Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0,10,10,30])
plt.legend()
plt.show()
```

### Bar Chart

```python
x1 = [1,2,3,4,5]
y1 = [10,11,17,12,16]


x2 = [6,7,8]
y2 = [20,22,24]

plt.bar(x1,y1,color='g',label='green', align='center')
plt.bar(x2,y2,color='b',label='blue', align='center')

plt.title('Bar Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0,10,8,30])
plt.legend()
plt.show()
```

### Horizontal Bar Chart

```python
x1 = [1,2,3,4,5]
y1 = [10,11,17,12,16]


x2 = [6,7,8]
y2 = [20,22,24]

plt.barh(y1,x1,color='g',label='green', align='center')
plt.barh(y2,x2,color='b',label='blue', align='center')

plt.title('Bar Chart')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0,10,8,30])
plt.legend()
plt.grid(True)
plt.show()
```

### Histograms

```python
from numpy.random import normal
gaussian_numbers = normal(size=1000)

plt.hist(gaussian_numbers)  # default bins = 10

plt.title('Histogram with Default Bins')
plt.xlabel('Value')
plt.ylabel('Freq')
plt.show()
```

```python
plt.hist(gaussian_numbers, bins = 20)

plt.title('Histogram with Specific No. of Bins')
plt.xlabel('Value')
plt.ylabel('Freq')
plt.show()
```

```python
plt.hist(gaussian_numbers, bins = 20, normed=True)  # Probability distribution

# matplotlib integrate the total area of the histogram and scale the values appropriately so that rather than showing 
# how many numbers in each bin, we instead have a probability of finding a number in that bin.
plt.title('Histogram with Probability Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')

plt.show()
```

### BoxPlots with Outliers

```python
x = [-1,0,1,1.2,1.3,-1.2,-1.4, 5] 
y = [-2,-1,0,1,2,-1.3,-0.5,0.7,1.5,6,6.5]

plt.axis([0,2,-4,7])

plt.boxplot([x,y], flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 5})

plt.show()
```

## Multiple Plots in the Same Figure

```python
x1 = [1,2,3,4,5]
y1 = [10,11,17,12,16]


x2 = [5,6,7]
y2 = [14,17,19]
x2_label = ['Five', 'Six', 'Seven']

plt.figure(1, figsize=(10,8)) # Sizing the figure so that all subplots are properly visible

plt.subplot(2,1,1)  # no. of rows, no. of columns, figure number
plt.xlim([0,8])
plt.ylim([9,20])

plt.plot(x1,y1, 'r--')

plt.subplot(2,1,2)
plt.xlim([0,8])
plt.ylim([9,20])
plt.xticks(x2, x2_label)  # Bundling labels and Numeric equivalents to display only labels
plt.plot(x2,y2, 'b--')


plt.show()
```
