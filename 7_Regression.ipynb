### Data Preprocessing

```python
# Preprocessing - Ordinal variables

degree_mapping = {'PhD': 4,
                'Master': 3,
                'Bachelor': 2,
                'High_School': 1}

level_mapping = {'IC': 1,
                'MM': 2,
                'D': 3,
                'VP': 4,
                'E': 5,
                'CEO': 6}

df['degree_level'] = df['degree_level'].map(degree_mapping)
df['level'] = df['level'].map(level_mapping)
```

```python
# Preprocessing - Nominal Categorical variables

# Create a list of features to dummy
todummy_list = ['signing_bonus', 'sex', 'dept']

# Function to dummy all the categorical variables used for modeling
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
    return df


df = dummy_df(df, todummy_list)
df.head()
```

### Partitioning into Train/Test Split

```python
# In reality, this step should be done just after preliminary data exploration


# Method 1 - Extract numpy matrix from dataframe

# display(df.shape)
# X, y = df.iloc[:, 1:15].values, df.iloc[:, 0:1].values
# y = y.ravel() # required to convert to 1-d array
# print "Before Standardizing"
# display(X)
# display(y)

# Method 2 - Extract using dataframe

y = df['salary']
X = df.drop('salary', axis = 1)

# Partition into train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # 80/20 split

# Check distribution after train/test split (for e.g. on column 'level'), if needed apply other ways of partitioning
print X_train.level.value_counts()
print X_test.level.value_counts()

# train/test split - stratify param, if need to split on specific criteria 
# Note: if proportion of train/test split cannot accomodate integer no. of obsv for criteria, will throw error
# cross_val_score and GridSearchCV - does stratified sampling by class labels in case of classifiers (by default)

# Standardize
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)   # fit + transform
X_test_std = stdsc.transform(X_test)         # only transform

print "\nAfter Standardizing\n"
display(X_train_std)
display(y_train) # indices will also appear
display(X_test_std)
display(y_test)

# Check models with and without standarized values 
# In general for linear regression it OK to not standardize
# If standardizing, interpretation gets tedious for continuous as well as categorical variables
# Standardized coefficients refer to how many standard deviations a dependent variable will change, per standard deviation increase in the predictor variable
# Standarization needed before Lasso : Standardization of the coefficient is usually done to answer the question of which of the independent variables have a greater effect on the dependent variable in a multiple regression analysis, when the variables are measured in different units of measurement
# For non- linear models, it is OK to standardize as we cannot interpret them as unit change in predictor results in what quantitative change in outcome variable
```

```python
def regression_linear(X_train,y_train,X_test,y_test,target):
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, median_absolute_error
    import matplotlib.pyplot as plt
    
    reg = LinearRegression()
    reg.fit(X_train, y_train)
    print "Training Error"
    print "R2 score = ", r2_score(y_train, reg.predict(X_train))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_train, reg.predict(X_train)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_train, reg.predict(X_train))
    print "Median Abs Error = ", median_absolute_error(y_train, reg.predict(X_train))
    print 

    # Coefficients of linear Regression, Note: Coeff will change if you standardize
    print "\n Multiple Linear Regression Model \n", sorted(zip(df.columns[df.columns != target],map(lambda x: round(x, 6), reg.coef_)))


    # Plot residuals - to check whether there is pattern or not
    print "\nResidual Plot\n"
    
    y_test_pred = reg.predict(X_test)
    plt.scatter(y_test_pred,  y_test_pred - y_test,
            c='black', marker='s', label='Test data')
    plt.xlabel('Predicted values')
    plt.ylabel('Residuals')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()
    
    print "Test Error"
    print "R2 score = ", r2_score(y_test, reg.predict(X_test))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_test, reg.predict(X_test)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_test, reg.predict(X_test))
    print "Median Abs Error = ", median_absolute_error(y_test, reg.predict(X_test))

regression_linear(X_train,y_train,X_test,y_test,'salary')
```

#### Random Forest Regression

```python
def regression_randomForest(X_train, y_train, X_test, y_test, target, scoring_param='neg_mean_squared_error'):
    from sklearn.model_selection import GridSearchCV
    from sklearn.ensemble import RandomForestRegressor
    import matplotlib.pyplot as plt
    import numpy as np
    
    rf = RandomForestRegressor()
    print rf.get_params()
    
    estimator_range = [10, 100, 1000]
    feature_range = ['sqrt', 'log2']
    depth_range = [3, 4]
    #Overall 3*2*2 = 12 combinations with five times CV

    param_grid = {'n_estimators': estimator_range, 'max_features': feature_range, 'max_depth' = depth_range}

    gs = GridSearchCV(estimator=rf, param_grid=param_grid, scoring=scoring_param, cv=5, n_jobs=-1)

    gs.fit(X_train,y_train)

    print "\nBest Score = {}".format(np.sqrt(-gs.best_score_))
    print "\nBest params: {}".format(gs.best_params_)
    print "\nBest Estimator: {}".format(gs.best_estimator_)
    
    rf = gs.best_estimator_
    
    feature_importance = rf.feature_importances_
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    pos = np.arange(sorted_idx.shape[0]) + .5
    plt.figure(figsize=(16, 12))
    plt.barh(pos, feature_importance[sorted_idx], align='center', color='#7A68A6')
    plt.yticks(pos, np.asanyarray(df.columns[df.columns != target])[sorted_idx])
    plt.xlabel('Relative Importance')
    plt.title('Variable Importance')
    plt.show()
    
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, median_absolute_error
    print "Training Error"
    print "R2 score = ", r2_score(y_train, rf.predict(X_train))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_train, rf.predict(X_train)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_train, rf.predict(X_train))
    print "Median Abs Error = ", median_absolute_error(y_train, rf.predict(X_train))
    print 
    print "Test Error"
    print "R2 score = ", r2_score(y_test, rf.predict(X_test))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_test, rf.predict(X_test)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_test, rf.predict(X_test))
    print "Median Abs Error = ", median_absolute_error(y_test, rf.predict(X_test))
    
regression_randomForest(X_train, y_train, X_test, y_test, 'salary')
```

#### Gradient Boosting Regression

```python
def regression_gradientBoosting(X_train, y_train, X_test, y_test, target, scoring_param='neg_mean_squared_error'):
    from sklearn.model_selection import GridSearchCV
    from sklearn.ensemble import GradientBoostingRegressor
    import matplotlib.pyplot as plt
    import numpy as np
    
    gb = GradientBoostingRegressor()
    print gb.get_params()
    
    estimator_range = [10, 100, 1000]
    learning_rate_range = [0.1, 0.2]
    max_depth_range = [3, 5, 7]
    feature_range = ['sqrt', 'log2']

    param_grid = {'n_estimators': estimator_range, 'learning_rate': learning_rate_range, 'max_depth': max_depth_range, 'max_features': feature_range}

    gs = GridSearchCV(estimator=gb, param_grid=param_grid, scoring=scoring_param, cv=2, n_jobs=-1)
    gs.fit(X_train,y_train)
    print "\nBest Score = {}".format(np.sqrt(-gs.best_score_))
    print "\nBest params: {}".format(gs.best_params_)
    print "\nBest Estimator: {}".format(gs.best_estimator_)
    
    gb = gs.best_estimator_
    
    feature_importance = gb.feature_importances_
    feature_importance = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    pos = np.arange(sorted_idx.shape[0]) + .5
    plt.figure(figsize=(16, 12))
    plt.barh(pos, feature_importance[sorted_idx], align='center', color='#7A68A6')
    plt.yticks(pos, np.asanyarray(df.columns[df.columns != target])[sorted_idx])
    plt.xlabel('Relative Importance')
    plt.title('Variable Importance')
    plt.show()
    
    from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, median_absolute_error   
    print "Training Error"
    print "R2 score = ", r2_score(y_train, gb.predict(X_train))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_train, gb.predict(X_train)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_train, gb.predict(X_train))
    print "Median Abs Error = ", median_absolute_error(y_train, gb.predict(X_train))
    print 
    print "Test Error"
    print "R2 score = ", r2_score(y_test, gb.predict(X_test))
    print "Root Mean Squared Error = ", np.sqrt(mean_squared_error(y_test, gb.predict(X_test)))
    # To check effect of outliers
    print "Mean Abs Error = ", mean_absolute_error(y_test, gb.predict(X_test))
    print "Median Abs Error = ", median_absolute_error(y_test, gb.predict(X_test))
    return gb
    
gb = regression_gradientBoosting(X_train, y_train, X_test, y_test, 'salary')
# Returning estimator is required for ploting partial dependence plots    
```

```python
# Partial dependence plot for interpretation of the model (defined only on training data)

from sklearn.ensemble.partial_dependence import plot_partial_dependence

fig, axs = plot_partial_dependence(gb, X_train, features = [1,3]) # The features are the column numbers from the X array 

# One-way PDPs tell us about the interaction between the target response and the target feature (e.g. linear, non-linear)
# Also it can tell direction ( +ve or -ve relationship)
```

##### Analyzing Error & Re-iterating - Some tips

* Outliers - Remove them or apply algorithms robust to outliers e.g. RANSAC
* Interaction terms - Polynomial regression
* Feature Selection - e.g. using Randomized Lasso, Feature Extraction - e.g. PCA , Feature Transformation
* Cluster and then apply different regression model for each cluster
* Ensemble Technique
* If missing values were large initially, analyze the effect of imputing or apply other imputation techniques
