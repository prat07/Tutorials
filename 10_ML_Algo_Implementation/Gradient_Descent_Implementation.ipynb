```python
import numpy as np
import random
```

```python
def gen_data(numpoints, bias, variance):
    x = np.zeros((numpoints, 2))
    y = np.zeros(numpoints)
    for i in range (0, numpoints):
        # bias feature
        x[i][0] = 1
        x[i][1] = i
        # target variable
        y[i] = i + bias + random.uniform(0,1) * variance
    return x, y
        

    
    
    
    
gen_data(10,25,10)
```

```python
# This is batch gradient descent
def gradient_descent(X, Y, alpha, numIter):
    n, m = np.shape(X)
    theta = np.ones(m)
    for i in range(0, numIter):
        hypothesis = np.dot(X, theta) # Activation function e.g. sigmoid can be applied to hypothesis after dot product 
        error = hypothesis - Y  
        # avg cost per observation
        cost = np.sum(error ** 2)/ (2*n)
        print("Iteration %d, Cost %f" %(i,cost))
        # avg gradient per observation
        gradient = np.dot(X.T, error) / n
        # update
        theta = theta - alpha * gradient
        print theta
    return theta
        
```

```python
X, Y = gen_data(10,25,10)
alpha = 0.005
numIter = 100

print gradient_descent(X, Y, alpha, numIter)
```
